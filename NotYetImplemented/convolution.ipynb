## Convolution and Input Shaping

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
```

```python
omega = 72
T_end = 0.1390
n = 1
num=30000
dt = T_end/num
dt
```

```python
input = np.hstack((np.linspace(0,0,1),np.linspace(1,1,num),np.linspace(0,0,1)))
t = np.linspace(0.0,dt*num,len(input))

fig=plt.figure(figsize=(8, 3))
ax=fig.add_axes([0,0,1,1])
ax.plot(t,input)
ax.grid()
```

```python
t1 = 0.0
t2 = n*np.pi/omega
A1 = 0.5
A2 = 0.5

impulse = np.linspace(0.0, 0.0,  len(input))
idx_t2 = np.abs(t - t2).argmin()
impulse[0] = A1
impulse[idx_t2] = A2
impulse = impulse[0:idx_t2+1]

fig=plt.figure(figsize=(8, 3))
ax=fig.add_axes([0,0,1,1])
ax.plot(np.linspace(0,dt*len(impulse),len(impulse)),impulse)
ax.grid()

t2, idx_t2
```

```python
input_shaped = np.convolve(input,impulse)
# idx_begin = input_shaped.nonzero()[0][0]-1

# input_shaped = input_shaped[idx_begin:len(input_shaped)]
t_new = np.linspace(0,dt*len(input_shaped), len(input_shaped))

fig=plt.figure(figsize=(8, 3))
ax=fig.add_axes([0,0,1,1])
ax.plot(t_new, input_shaped)
ax.set_title('Input shaped trajectory')
ax.set_xlabel('time in sec')
ax.set_ylabel('amplitude')
ax.grid()

T_end+t2, idx_begin
```
